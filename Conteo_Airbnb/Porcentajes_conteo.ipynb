```python
import pandas as pd


data = pd.read_csv('porcentaje_habitaciones_precio_comunidad.csv')


data['suma_total'] = data['Conteo_limpio'] + data['habitaciones'] + data['descuento']


data['porcentaje_habitaciones'] = (data['habitaciones'] / data['suma_total']) * 100


data['porcentaje_descuento'] = (data['descuento'] / data['suma_total']) * 100


data.to_csv('resultado_con_porcentajes_total.csv', index=False)


print(data)
```
