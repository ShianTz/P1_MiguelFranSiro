```python
import numpy as np
import pandas as pd
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Leemos el archivo CSV con la info de AirBNB
infoAirBNB = "info_total_airbnb_7.csv"
df = pd.read_csv(infoAirBNB)

df_limpio = df.drop("url", axis=1)

df_limpio.rename(columns={'tipo_alojamineto': 'tipo_alojamiento'}, inplace=True)

df_limpio.info()
```

```python
from ydata_profiling import ProfileReport
path_2 = '/Users/migueldelhoyo/Programacion/Hack_a_boss/Apuntes/'
report = ProfileReport(df_limpio, title='Report AirBNB')
report.to_file(output_file= path_2 + "AirBNB.html")
```

```python
#Cruce de variables

sns.pairplot(df_limpio)
plt.show()
```

```python
# Cruzamos variables por el tipo de alojamiento

sns.pairplot(df_limpio, hue = "tipo_alojamiento")

plt.savefig("Variables_por_tipo_alojamiento.png")
plt.show()
```

```python
#Histograma del precio

plt.figure(figsize=(8, 6))
sns.histplot(df_limpio['precio'], kde=True)
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.title('Histograma del Precio')

plt.savefig("Precio.png")
plt.show()
```

```python
#Grfico de dispersin precio vs valoracin

plt.figure(figsize=(8, 6))
sns.scatterplot(x='precio', y='valoracion', data=df_limpio)
plt.xlabel('Precio')
plt.ylabel('Valoracin')

plt.savefig("Precio_vs_Valoracin.png")
plt.show()
```

```python
# Grfico de dispersin valoracin vs nmero de habitaciones

plt.figure(figsize=(8, 6))
sns.scatterplot(x='n_dormitorios', y='valoracion', data=df_limpio)
plt.xlabel('Nmero de Habitaciones')
plt.ylabel('Valoracin')
plt.title('Valoracin vs Nmero de Habitaciones')
plt.show()
```

```python
# Grfico de dispersin valoracin vs nmero de baos

plt.figure(figsize=(8, 6))
sns.scatterplot(x='n_banos', y='valoracion', data=df_limpio)
plt.xlabel('Nmero de Baos')
plt.ylabel('Valoracin')
plt.title('Valoracin vs Nmero de Baos')

plt.savefig("Baos_vs_Valoracin.png")
plt.show()
```

```python
# Grfico de dispersin precio vs nmero de baos

plt.figure(figsize=(8, 6))
sns.scatterplot(x='n_banos', y='precio', data=df_limpio)
plt.xlabel('Nmero de Baos')
plt.ylabel('Precio')
plt.title('Precio vs Nmero de Baos')
plt.show()
```

```python
# Grfico de cajas y bigotes tipo de alojamiento vs valoracin

plt.figure(figsize=(10, 6))
sns.boxplot(x='tipo_alojamiento', y='valoracion', data=df_limpio)
plt.xlabel('Tipo de Alojamiento')
plt.ylabel('Valoracin')
plt.title('Valoracin por Tipo de Alojamiento')
plt.xticks(rotation=45)

plt.savefig("Tipo_Alojamiento_vs_Valoracin.png")
plt.show()
```

```python
# Grfico de barras tipo de alojamiento vs comunidad

plt.figure(figsize=(12, 6))
sns.countplot(x='comunidad', hue='tipo_alojamiento', data=df_limpio)
plt.xlabel('Comunidad')
plt.ylabel('Cantidad de Alojamientos')
plt.title('Distribucin de Tipos de Alojamiento por Comunidad')
plt.xticks(rotation=90)
plt.legend(title='Tipo de Alojamiento')

plt.savefig("Tipo_Alojamiento_vs_Comunidad.png")
plt.show()
```

```python
# Grfico de barras precio vs comunidad

plt.figure(figsize=(10, 6))
sns.barplot(x='comunidad', y='precio', data=df_limpio, estimator=sum)
plt.xlabel('Comunidad')
plt.ylabel('Precio')
plt.title('Precio de Alojamientos por Comunidad')
plt.xticks(rotation=90)

plt.savefig("Precio_vs_Comunidad.png")
plt.show()
```

```python
# Grfico de barras valoracin promedio vs comunidad

plt.figure(figsize=(10, 6))
sns.barplot(x='comunidad', y='valoracion', data=df_limpio)
plt.xlabel('Comunidad')
plt.ylabel('Valoracin promedio')
plt.title('Valoracin Promedio por Comunidad')
plt.xticks(rotation=90)

plt.savefig("Valoracin_vs_Comunidad.png")
plt.show()
```
